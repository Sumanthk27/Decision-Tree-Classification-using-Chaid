```python
# Define a function for calculating chi-square value
def chi_square(observed, expected):
    return np.sum((observed-expected)**2/expected)
```

```python
# Define a function to find the best split feature
def find_best_feature(data, target="Churn"):
    best_feature = None
    best_chi_square = 0
    
    for feature in data.columns:
        if feature == target:
            continue
            
        observed = pd.crosstab(data[feature], data[target]).values
        feature_sum = np.sum(observed, axis=1)
        target_sum = np.sum(observed, axis=0)
        total_sum = np.sum(feature_sum)
        expected = np.outer(feature_sum, target_sum) / total_sum
        
        feature_chi_square = chi_square(observed, expected)
        
        if feature_chi_square > best_chi_square:
            best_chi_square = feature_chi_square
            best_feature = feature

        if best_feature is None:
           return target
    
    return best_feature
```

```python
def build_tree(data, target, features, max_depth=None, current_depth=0, min_samples_split=2, min_samples_leaf=1, random_state=None):
    # Base case 1: maximum depth is reached
    if max_depth and current_depth == max_depth:
        return {'Churn': data[target].mode()[0]}
    
    # Base case 2: no features left
    if len(features) == 0:
        return {'Churn': data[target].mode()[0]}
    
    # Find the best feature to split the data
    best_feature = find_best_feature(data, target)
    
    # Create a new node with the best feature
    root = {'feature': best_feature, 'Churn': data[target].mode()[0], 'children': {}}
    
    # Remove the best feature from the list of features
    features = [f for f in features if f != best_feature]

    # Check if the node should be split further
    if len(data) < min_samples_split:
        return {'Churn': data[target].mode()[0]}
    
    # Split the data by the best feature
    for value in np.unique(data[best_feature]):
        sub_data = data[data[best_feature] == value]
        sub_tree = build_tree(sub_data, target, features, max_depth, current_depth+1,min_samples_split, min_samples_leaf, random_state)
        root['children'][value] = sub_tree

    # Add a 'Loan_Status' key at leaf nodes
    if len(root['children']) == 0:
        root['Churn'] = data[target].mode()[0]
        
    return root
```

```python
# Define a function to predict the class for a new instance
def predict(instance, tree):
    # Traverse the tree until a leaf node is found
    while 'feature' in tree:
        feature_value = instance[tree['feature']]
        if feature_value not in tree['children']:
            break
        tree = tree['children'][feature_value]
        
    # Return the predicted class
    return tree['Churn']
```

```python
# Build the decision tree using the train data
target = 'Churn'
features = list(train_data.columns[:-1])
tree = build_tree(train_data, target, features, max_depth=6, min_samples_split=3, min_samples_leaf=1)
```

```python
predictions = []
for i, instance in test_data.iterrows():
    predictions.append(predict(instance, tree))
```

```python
accuracy = np.mean(predictions == test_data[target])
print("Accuracy:", accuracy)
```

```python
def grid_search(train_data, test_data, target, features, max_depth_list, min_samples_split_list, min_samples_leaf_list, random_state=None):
    best_accuracy = 0
    best_tree = None
    best_params = None
    
    for max_depth in max_depth_list:
        for min_samples_split in min_samples_split_list:
            for min_samples_leaf in min_samples_leaf_list:
                tree = build_tree(train_data, target, features, max_depth=max_depth, min_samples_split=min_samples_split, min_samples_leaf=min_samples_leaf, random_state=random_state)

                # Calculate the accuracy of the tree on the test data
                predictions = [predict(instance, tree) for i, instance in test_data.iterrows()]
                accuracy = np.mean(predictions == test_data[target])

                if accuracy > best_accuracy:
                    best_accuracy = accuracy
                    best_tree = tree
                    best_params = {'max_depth': max_depth, 'min_samples_split': min_samples_split, 'min_samples_leaf': min_samples_leaf}
    
    return best_accuracy, best_tree, best_params
```

```python
# Define the hyperparameters to search over
max_depth_list = [3, 4, 5, 6]
min_samples_split_list = [2, 3, 4, 5]
min_samples_leaf_list = [1, 2, 3, 4]
```

```python
target = "Churn"
features = [f for f in train_data.columns if f != target]
best_accuracy, best_tree, best_params = grid_search(train_data, test_data, target, features, max_depth_list, min_samples_split_list, min_samples_leaf_list, random_state=42)
```
